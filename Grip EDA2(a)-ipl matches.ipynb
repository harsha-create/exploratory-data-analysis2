```python
g= sns.FacetGrid(df,row="team1",col="winner",hue="winner",palette="Set1")
kws = dict(s=50,linewidth=0.5,edgecolor="black")
g= g.map(plt.scatter,"win_by_runs","win_by_wickets",**kws)
g.set(xlim=(0,80),ylim=(0,30))
for ax in g.axes.flat:
    ax.plot((0,80),(0,30),c="orange",ls="--")
g.add_legend()
```
